```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/weight-height.csv')
df.head()
```

```python
df.corr()
```

```python
x = df['Height']
y = df['Weight']
x = x.to_frame()
y = y.to_frame()
```

```python
pip install statsmodels
```

```python
import statsmodels.api as sm
```

```python
pip install sklearn
```

```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
lm
```

```python
model_lm = lm.fit(y,x)
```

```python
print('La interpeción es \t:',model_lm.intercept_)
print('Los coeficientes son \t:',model_lm.coef_)
print('El valor de R^2 es \t:',model_lm.score(y,x))
```

```python
model_ols = sm.OLS.from_formula('x~y', data =df) #Height = bo + b1*Weight
result = model_ols.fit()
result.summary()
```

```python
numeros = ([200,201,94])
numeros = pd.DataFrame(numeros)
numeros
```

```python
y = model_lm.predict(numeros)
y = pd.DataFrame(y)
y
```

```python
df = pd.read_csv('datasets/nhanes_2015_2016.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df.columns
```

```python
vars = ['SMQ020', 'RIAGENDR', 'RIDAGEYR','BMXWT', 'BMXHT', 'BMXBMI', 'BMXWAIST', 'HIQ210']
```

```python
df = df[vars].dropna()
df
```

```python
df.corr()
```

#### BMXWT = bo + b1*RIDAGEYR

```python
model = sm.OLS.from_formula('BMXWT~RIDAGEYR', data =df)
result = model.fit()
result.summary()
```

```python
df['RiaGENDR'] = df.RIAGENDR.replace({1:'Male', 2:'Female'})
df
```

#### BMXWT = bo + b1*RIDAGEYR + b2*RIAGENDR
#### BMXWT = bo + b1*RIDAGEYR + b2(Male+Female)
#### BMXWT = 77.5710 - 0.0228*RIDAGEYR + 10.5371*Male

```python
model = sm.OLS.from_formula('BMXWT~RIDAGEYR+RiaGENDR', data =df)
result = model.fit()
result.summary()
```
