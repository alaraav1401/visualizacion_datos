#### Pregunta de investigación 1
##### ¿Existe alguna correlación entre Wingspan y CWDistance?

```python
sns.lmplot(x = 'Wingspan',
           y = 'CWDistance',
           data = df,
           #fit_reg = False
           #hue = 'Gender'
          )

plt.title('Scatterplot of Wingspan vs CWDistance', fontsize=16)
```

#### Conclusión
