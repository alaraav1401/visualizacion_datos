```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%ls datasets
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

### Pregunta 1

```python
sns.boxplot(x = df['peso'],
           whis = (0,100))
```

### Pregunta 2

```python
pivote2 = pd.pivot_table(df,
                          index = 'desc_ent',
                          values = 'edad',
                          columns = 'sexo',
                          aggfunc = 'median',
                         )
df2 = pivote2
df2.head()
```

```python
sns.boxplot(data = df2)
```

### Pregunta 3

```python
pivote3 = pd.pivot_table(df,
                          index = 'desc_ent',
                          values = 'cadera',
                          columns = 'sexo',
                          aggfunc = 'median',
                         )
df3 = pivote3
df3.head()
```

```python
df3.plot()
```

### Pregunta 4

```python
pivote4 = pd.pivot_table(df,
                          index = 'desc_ent',
                          values = 'ropa',
                          columns = 'sexo',
                          aggfunc = 'sum',
                         )
df4 = pivote4
df4.head()
```

```python
df4_top10 = df4.sort_values(by = 2,
                         ascending = False,
                         axis = 0)
df4_top10 = df4_top10.head(10)
df4_top10.drop([1], inplace=True, axis=1)
df4_top10
```

```python
df4_top10.plot(kind='barh', figsize=(15,8))
```
