```python
import pandas as pd
```

```python
df = pd.read_excel('datasets/Canada.xlsx',
                   sheet_name = 'Canada by Citizenship',
                   engine = 'openpyxl',
                   skiprows = range(20),
                   skipfooter = 2 #para saltarse los últimos 2 renglones
                  )
#df.head()
```

```python
df.columns
```

```python
all(isinstance(column, str) for column in df.columns)
```

```python
df.columns = list(map(str,df.columns))
all(isinstance(column, str) for column in df.columns)
```

```python
df.drop(['Type','Coverage','REG','DEV','DevName','Unnamed: 43','Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47',
         'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'], inplace=True, axis=1)
df.columns
```

```python
df.rename(columns = {'OdName': 'Country',
                     'AreaName': 'Continent',
                     'RegName': 'Region',
                    },
          inplace = True)
df.columns
```

```python
df['Total'] = df.iloc[:,4:].sum(axis=1)
df.set_index('Country', inplace=True)
df.head()
```

### Pregunta de investigación #1
#### Me gustaría saber cuantas personas llegaron desde Haití a Canada desde 1980-2013, con una gráfica de líneas

```python
df.loc['Haiti','2011'].sum()
```

```python
df.loc['Haiti', years].plot()
plt.xlabel('Año',fontsize=16)
plt.xticks(fontsize=16, rotation=45)
plt.ylabel('Número de inmigrantes',fontsize=16)
plt.yticks(fontsize=16, rotation=45)
plt.title('Número de inmigrantes desde Haití a Canada de 1980-2013', fontsize=13)
plt.text(31,  6503, 'terremoto')
plt.plot(31,  6503, 'o')
plt.show()
```

```python
dfN = df.loc[['China','India'],years]
dfN = dfN.transpose()
dfN.head(2)
```

```python
dfN.plot(kind='line')
```

```python
df_top5 = df.sort_values(by = 'Total',
                         ascending = False,
                         axis = 0)
df_top5 = df_top5.head(7)
df_top5.drop(['Total','Unknown'], inplace=True, axis=0)
df_top5 = df_top5.loc[:,years]
df_top5 = df_top5.transpose()
df_top5.plot(kind='area', figsize=(15,8), alpha = 0.5)
```
