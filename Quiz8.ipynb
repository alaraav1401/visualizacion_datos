#### ¿Cual es la histograma del  diastol  de siguiente graficas (bins = 15) ?

```python
#pip install seaborn
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
df.columns
```

```python
sns.histplot(df['diastol'],
             kde = False,
            bins = 15)
```

#### ¿Cual seria diagrama de dispersión cuando x='sistol',  y='cintura',  hue='sexo'?

```python
sns.lmplot(x = 'sistol',
           y = 'cintura',
           data = df,
           fit_reg = False,
           hue = 'sexo'
          )
```

#### Si 'sexo 2 = mujer', y 'sexo 1 = hombre', no hay valores atípicos en mujer_cintura

```python
sns.boxplot(data = df.loc[df['sexo'] == 2,['cintura']])
```

#### Si 'sexo 2 = mujer', y 'sexo 1 = hombre', cual seria diagrama dispersion cuando x = hombre_sistol,  y = hombre_diastol

```python
df2 = df.loc[df['sexo']==1, ['sistol','diastol']]
df2.head()
```

```python
sns.lmplot(x = 'sistol',
           y = 'diastol',
           data = df2,
           fit_reg = False,
          )
```

#### ¿Si 'sexo 2 = mujer', y 'sexo 1 = hombre', cual seria histograma de  mujer_peso?

```python
df3 = df.loc[df['sexo']==2, ['peso']]
df3.head()
```

```python
sns.histplot(df3['peso'],
             kde = False,
            bins = 25
            )
```
